```python
# I would like to see if a multi-scale Harris corner detector is achievable.
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import scipy.signal
import scipy.ndimage
import scipy.ndimage.filters as filters

def load_image(filepath):
    """Loads an image into a numpy array.
    Note: image will have 3 color channels [r, g, b]."""
    img = Image.open(filepath)
    return (np.asarray(img).astype(np.float)/255)[:, :, :3]

def get_local_maxima(data, threshold):
    # See: https://stackoverflow.com/a/9113227/3672986
    neighborhood_size = 3

    data_region_max = filters.maximum_filter(data, neighborhood_size)
    maxima = (data == data_region_max)
    data_min = filters.minimum_filter(data, neighborhood_size)
    maxima[data < threshold] = 0

    labeled, num_objects = scipy.ndimage.label(maxima)
    slices = scipy.ndimage.find_objects(labeled)
    x, y = [], []
    for dy,dx in slices:
        x_center = int(round((dx.start + dx.stop - 1)/2))
        x.append(x_center)
        y_center = int(round((dy.start + dy.stop - 1)/2))   
        y.append(y_center)
        
    return np.array(list(zip(x, y)))

def get_local_maxima_3(data, threshold, sigmas):
    # See: https://stackoverflow.com/a/9113227/3672986
    neighborhood_size = 3

    data_region_max = filters.maximum_filter(data, neighborhood_size)
    maxima = (data == data_region_max)
    data_min = filters.minimum_filter(data, neighborhood_size)
    maxima[data < threshold] = 0

    labeled, num_objects = scipy.ndimage.label(maxima)
    slices = scipy.ndimage.find_objects(labeled)
    
    peaks = 
    x, y = [], []
    for dy,dx in slices:
        x_center = int(round((dx.start + dx.stop - 1)/2))
        x.append(x_center)
        y_center = int(round((dy.start + dy.stop - 1)/2))   
        y.append(y_center)
        
    return np.array(list(zip(x, y)))


def get_gaussian_filter(n, sigma=1):
    assert(n % 2 == 1)
    h = (n + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d = d[:, np.newaxis]
    d_sq = d**2 + d.T ** 2
    # Take the gaussian
    g = np.exp(-d_sq/2/(sigma**2))
    # Normalize
    g = g/g.sum().sum()
    return g

def get_harris_components(image):
    sobel_x = np.array([
      [1, 0, -1],
      [2, 0, -2],
      [1, 0, -1]
    ])

    sobel_y = np.array([
      [1, 2, 1],
      [0, 0, 0],
      [-1, -2, -1]
    ])

    Ix = scipy.signal.convolve(
        image, sobel_x, mode='same')
    Iy = scipy.signal.convolve(
        image, sobel_y, mode='same')

    # Compute the Harris operator matrix
    w = get_gaussian_filter(15, sigma=3)
    A = scipy.signal.convolve(
        Ix * Ix, w, mode='same')
    B = scipy.signal.convolve(
        Ix * Iy, w, mode='same')
    C = scipy.signal.convolve(
        Iy * Iy, w, mode='same')
    return (A, B, C)

def get_harris_corners(image, sigma=1.0):
    A, B, C = get_harris_components(image)
    det = A*C - B*B
    tr = A + C
    f = det/tr * (sigma**2)
    return get_local_maxima(f, threshold=0.05)
```

```python
# Feature response as a function of sigma
image_base = load_image("light_cubes_sm.png")[:, :, 0]
corners = get_harris_corners(image)


sigmas = np.arange(1.0, 32.0, 0.5)
response = np.zeros(
    [image_base.shape[0], image_base.shape[1], sigmas.size]
)

for ii, sigma in enumerate(sigmas):
    w = get_gaussian_filter(101, sigma=sigma)
    image = scipy.signal.convolve(
        image_base, w, mode='same')
    # Harris features
    A, B, C = get_harris_components(image)
    det = A*C - B*B
    tr = A + C
    response[:, :, ii] = det/tr * (sigma**2)

fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(1, 2, 1)
ax.imshow(image_base)
ax = plt.subplot(1, 2, 2)
ax.imshow(np.argmax(response, axis=2))
```
