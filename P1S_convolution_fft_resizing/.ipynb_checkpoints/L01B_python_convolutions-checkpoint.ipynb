```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt

def convolve_1(signal, filt):
    out = np.zeros_like(signal)
    for iS in range(filt.shape[0]-1, signal.shape[0]):
        for iF in range(filt.shape[0]):
            out[iS] += signal[iS-iF] * filt[iF]
    # Remove entries in 'out' that were not set
    out = out[filt.shape[0]-1:]
    return out

def convolve_2(signal, filt):
    out = np.zeros_like(signal)
    fsize = filt.shape[0]
    # The convolution flips the filter before applying it
    filt_reversed = np.flip(filt)
    for iS in range(signal.shape[0]-fsize+1):
        # We can use the dot product to compute the output
        out[iS] = np.dot(signal[iS:iS+fsize], filt_reversed)
    # Remove entries in 'out' that were not set
    out = out[:1-filt.shape[0]]
    return out

def convolve_3(sigal, filt):
    return scipy.signal.convolve(signal, filt, mode='valid')

    
def get_gaussian_filter(filter_width, sigma):
    raise NotImplementedError()

def get_laplace_of_gaussian_filter(filter_width, sigma):
    raise NotImplementedError()
    
def compute_example_signal(length=256, step_width=64):
    step_center = length//2
    x = np.arange(length) - step_center
    signal = (np.abs(x) < step_width/2).astype(np.float)
    # Add noise to the signal
    signal += np.random.normal(scale=0.05, size=signal.shape)
    return signal
    
plt.figure(dpi=300)
signal = compute_example_signal()
box_filter_1d = np.ones(11)/11
plt.subplot(1, 2, 1)
plt.plot(signal)
plt.subplot(1, 2, 2)
csignal = convolve_1(signal, box_filter_1d)
plt.plot(csignal)
csignal = convolve_2(signal, box_filter_1d)
plt.plot(csignal)
csignal = convolve_3(signal, box_filter_1d)
plt.plot(csignal)

None
```
