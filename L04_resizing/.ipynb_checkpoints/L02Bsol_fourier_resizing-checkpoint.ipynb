## L02B.1 Computing Derivatives with Fourier Transforms

In class, we discussed computing the Fourier Transform of a signal, in which we could represent any periodic signal as a sum of sine and cosine functions (or complex exponentials):

$$ f(x) = \sum_{k = 0}^{N - 1} F(k) e^{i 2 \pi k x / N} $$

Computing the derivative of a Fourier Transformed signal is relatively straightforward:

$$ \frac{d}{dx}f(x) = \sum_{k = 0}^{N - 1} \frac{i 2 \pi k}{N}F(k) e^{i 2 \pi k x / N} $$

We can compute derivatives by multiplying in frequency space. This means that taking the Fourier Transform, multiplying by a frequency-dependent term, and taking the inverse Fourier Transform can be used to compute the derivative.In practice, the process is still relatively simple, but there are some pitfals you should know to avoid.

For this breakout session, I have provided you with some starter code for a few signals you should know the derivative of. Use the theorem above, along with the [numpy Fast Fourier Transform documentation](https://numpy.org/doc/stable/reference/routines.fft.html), to **compute the derivative of this function**.

Steps:
- Compute the Fast Fourier Transform of each signal using numpy: `np.fft.fft`. Plot each signal and its frequency-space representation (on different axes). Keep in mind that the frequency-space representation results in a complex (both real and imaginary) signal (use `np.real`, `np.imag`, `np.abs` and plot all three on the same axis). What do you notice about the frequency-space representation? Where are the low-frequency terms? The high-frequency terms? (`np.fft.fftshift` is often used to manipulate the signal into a form that is easier to work with.)
- Compute the frequency-space vector `k` and use it to compute the derivative. Define `k` so that the derivative is accurate even if `L` and `N` are changed. Confirm that the derivative you compute with the Fourier Transform matches the analytic derivative by plotting them on top of one another. Note that your absolute scale may be off: `k` will depend on both `N` and `L`. (Do not use `np.fft.fftfreq`; it is instructive to do this by hand at least once.)

If you still have time, generate a square wave (or another "interesting" signal of your choosing) and apply a high-pass and low-pass filter in frequency space. What do you notice? How does the cutoff frequency impact the transform?

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Signals
L = 1.0
N = 101
x = np.arange(0, L, L / N)
signal_a = np.zeros(N, dtype=np.float)
signal_a[1*N//4:3*N//4] = 1.0
signal_b = np.sin(2 * np.pi * 4 * x / L)

# Plot the signals
fig = plt.figure(figsize=(8, 4))
ax = plt.subplot(1, 2, 1)
ax.plot(x, signal_a, '.')
ax.set_title('Signal A')

fig = plt.figure(figsize=(8, 4))
ax = plt.subplot(1, 2, 1)
ax.plot(x, signal_b, '.')
ax.set_title('Signal B')

def fft_derivative(signal_x, w):
    """Uses the Fourier Transform to compute the derivative
    of a function. Requires a signal and the 'frequency vector'
    w."""
    return np.fft.ifft(1j * w * np.fft.fft(signal_x))

raise NotImplementedError("Define 'w'")
d_signal_a = fft_derivative(signal_a, w)

None
```

```python
# Solution

L = 1.0
N = 101
x = np.arange(0, L, L / N)
signal_x = np.sin(2 * np.pi * 4 * x / L)
d_signal_x = 2 * np.pi * 4 / L * np.cos(2 * np.pi * 4 * x / L)
w = np.fft.fftfreq(N) * N * 2 * np.pi / L
d_signal_x_fft = np.fft.ifft(1j * w * np.fft.fft(signal_x))

# Plotting
fig = plt.figure(figsize=(8, 4), dpi=150)
ax = plt.subplot(1, 2, 1)
ax.plot(x, d_signal_x, '.')
ax.set_title('Analytic Derivative')

ax = plt.subplot(1, 2, 2)
ax.plot(x, np.real(d_signal_x_fft), '.')
ax.set_title('FFT-computed Derivative')

None
```

```python
# Some demo figures
import scipy.signal

def get_gaussian_filter_1D(n, sigma=1):
    assert(n % 2 == 1)
    h = (n + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d_sq = d**2
    # Take the gaussian
    g = np.exp(-d_sq/2/(sigma**2))
    # Normalize
    g = g/g.sum().sum()
    return g

def plot_signal_transformed(signal, filt, title):
    sig_f = scipy.signal.convolve(signal, filt, mode='same')
    sig_w = np.fft.fft(signal)
    sig_f_w = np.fft.fft(sig_f)
    
    plt.figure(dpi=150, figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(x, signal, 'b.')
    plt.plot(x, sig_f, 'rx')
    plt.title(title)
    
    plt.subplot(1, 2, 2)
    plt.plot(w, np.abs(sig_w), 'b.')
    plt.plot(w, np.abs(sig_f_w), 'rx')
    plt.title(title)
    
# Mean filter
identity_filter = np.array([1])
mean_filter_3 = np.ones((3))/3
mean_filter_15 = np.ones((15))/15
gaussian_filter_15 = get_gaussian_filter_1D(41, 5)
diff_mean_filter_15 = np.zeros(gaussian_filter_15.size)
diff_mean_filter_15[diff_mean_filter_15.size//2] = 1
diff_mean_filter_15 -= gaussian_filter_15

plot_signal_transformed(signal_a, identity_filter, 'Identity Filter')
plot_signal_transformed(signal_a, mean_filter_3, 'Mean Filter (3)')
plot_signal_transformed(signal_a, mean_filter_15, 'Mean Filter (15)')
plot_signal_transformed(signal_a, gaussian_filter_15, 'Gaussian Filter (5)')
plot_signal_transformed(signal_a, diff_mean_filter_15, '1 - Gaussian Filter (5)')
```

```python
from PIL import Image

def load_image(filepath):
    """Loads an image into a numpy array.
    Note: image will have 3 color channels [r, g, b]."""
    img = Image.open(filepath)
    return (np.asarray(img).astype(np.float)/255)[:, :, :3]

image = load_image("lake_long_exposure_scene.png")
image_r = image[:, :, 0]
image_g = image[:, :, 1]
image_b = image[:, :, 2]

fig = plt.figure(figsize=(8, 8), dpi=150)
ax = plt.subplot(2, 2, 1)
ax.imshow(image)
ax = plt.subplot(2, 2, 2)
ax.imshow(image_r, cmap='gray', vmin=0, vmax=1)
ax = plt.subplot(2, 2, 3)
ax.imshow(image_g, cmap='gray', vmin=0, vmax=1)
ax = plt.subplot(2, 2, 4)
ax.imshow(image_b, cmap='gray', vmin=0, vmax=1)

None
```
